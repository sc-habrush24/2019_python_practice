```python
#Homework Assignment 2

#Name: 최 솔

#Time Spent: 1:00

import random #random을 import하여 불러온다
import string #string을 import하여 불러온다.

WORDLIST_FILENAME = "words.txt"

def load_words(): #함수 load_words()를 정의한다
    """
    Returns a list of valid words. Words are strings of lowercase letters.
    
    Depending on the size of the word list, this function may
    take a while to finish.
    """
    print ("Loading word list from file...")
    # 위의 문장을 출력한다.
    inFile = open(WORDLIST_FILENAME, 'r')
    #inFile에 WORDLIST_FILENAME 을 열고 읽어낸다.
    line = inFile.readline()
    #line에 inFile을 한줄씩 읽어낸다.
    wordlist = line.split(" ")
    #wordlist 에 line에 split함수 사용한 것을 저장한다.
    print (len(wordlist), "words loaded.")
    #len(wordlist)와 "words loaded"를 출력한다
    return wordlist
    #wordlist를 반환한다.
    
def choose_word(wordlist): #wordlist안의 내용을 랜덤으로 출력하는 
                           #choose_word(wordlist)를 정의한다. 
    """
    wordlist (list): list of words (strings)

    Returns a word from wordlist at random
    """
    return random.choice(wordlist)

wordlist = load_words()#wordlist에 load_words()를 저장한다.
# your code begins here!


def hangman(): #hangman()함수를 정의한다.
    
    i=str(choose_word(wordlist))#i에 choose_word(wordlist)를 저장한다.
    
    m='_'*len(i)#m에 '-'*len(i)를 저장한다.
    
    chance_number=8 #chance_number를 8로 둔다.
    
    print('Welcome to the game, Hangman!\nI am thinking of a word that is',len(i),'letters long.')
    #'Welcome to the game, Hangman!\nI am thinking of a word that is',len(i),'letters long.'을 출력한다.
    
    n='abcdefghijklmnopqrstuvwxyz'
    #'abcdefghijklmnopqrstuvwxyz'을 n에 저장한다.
    
    while True : #while 반복문을 설정한다. 
        print('------------------------\nYou have',chance_number,'guesses left.\nAvailable letters :',n)
        #'------------------------\nYou have',chance_number,'guesses left.\nAvailable letters :',n 을 출력한다.
        while True: #while 반복문을 한번 더 설정한다.
            my_answer=input('Please guess a letter :') #my_answer를 인풋받는다.
            if my_answer in n and len(my_answer)==1: #만약 my_answer가 n에 해당하면서 길이가 1이면 반복문 탈출
                break
            print('Wrong input! Try again!')#'Wrong input! Try again!'을 출력한다.
        if my_answer in i : #만약 my_answer가 i에 해당한다면
            for k in range(len(i)): #for 문을 설정해서 range를 (len(i))로 두어
                if i[k]==my_answer: #만약 i[k]==my_answer면 m에 m[0:k]+my_answer+m[k+1:len(i)]를 저장한다.
                    m=m[0:k]+my_answer+m[k+1:len(i)]#그리고 'Good guess:',m를 출력한다
            print('Good guess:',m)
        else :
            print('Oops! That letter is not in my word:')#위의 if문에 해당하지 않는다면 
            chance_number-=1        #'Oops! That letter is not in my word:'를 출력하고 chance_number에서 -1을 한다.
        n=n.replace(my_answer,'')#n에서 my_answer를 지운 것을 다시 n에 저장한다.
        if chance_number==0 :#만약 chance_number가 0이면 'You Fail! The answer is',i를 출력하고 탈출한다.
            print('You Fail! The answer is',i)
            break
        elif m==i :#만약 m==i이면 'Congratulations, you won!'를 출력하고 탈출한다. 
            print('Congratulations, you won!')
            break
hangman()
            
```
