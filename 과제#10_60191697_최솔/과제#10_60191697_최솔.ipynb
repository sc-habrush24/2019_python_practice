```python
class Building(object):
    def __init__(self):        
        while True :
            try : 
                floor_number=int(input('input_num_of_floor :'))
                if floor_number <2 :
                    raise ValueError
                else :
                    self.floor_number=floor_number
                    break
            except:
                print('Wrong Input!!')
class Customer(Building):
    def __init__(self):
        Building.__init__(self)
        while True :
            try : 
                customer_number=int(input('input_num_of_customer :'))
                if customer_number <1 :
                    raise ValueError
                else :
                    self.customer_number=customer_number
                    break
            except :
                print("Wrong Input!!")
        self.customer_name=[]
        self.cur_list=[]
        self.dest_list=[]
        self.in_elevator=[]
        self.get_out=[]
        import random
        for i in range(self.customer_number):
            self.customer_name+=['customer'+str(i+1)]
            while True:
                cur=[random.randint(1,self.floor_number)]
                dest=[random.randint(1,self.floor_number)]
                if cur!=dest: break
            self.cur_list+=cur
            self.dest_list+=dest
            self.in_elevator+=[False]
            self.get_out+=[False]
class Elevator(Customer):
    def __init__(self):
        Customer.__init__(self)
        print()
        for k in range(self.floor_number):
            for i in range(self.customer_number):
                if self.cur_list[i]==k+1:
                    self.in_elevator[i]=True
                    print(self.customer_name[i],'->','Elevator')
                if self.in_elevator[i]==True:
                    if self.dest_list[i]==k+1:
                        self.get_out[i]=True
                        self.in_elevator[i]=False
                        print(self.customer_name[i],'-> Destination Floor')
            print("\nElevator's Current floor :",k+1,"       Elevator's direction : up")
            print("Customers:")
            for i in range(self.customer_number):
                print('       ',self.customer_name[i],'       current floor :',self.cur_list[i],'       destination floor :',self.dest_list[i])
                print('                 in Elevator :',self.in_elevator[i],'       is Finished:',self.get_out[i])
            print('\n------>------->-------->------->')
            if k<self.floor_number-1:
                print('\nElevator :',k+1,'->',k+2)
        print('\nThe elevator is on the Top Floor Now')
        for k in range(self.floor_number-1,min(self.dest_list)-1,-1):
            print('\nElevator : ',k+1,'-> ',k)
            for i in range(self.customer_number):
                if self.in_elevator[i]==True:
                    if self.dest_list[i]==k:
                        self.get_out[i]=True
                        self.in_elevator[i]=False
                        print('       ',self.customer_name[i], '->', 'Destination Floor')
            print("\nElevator's Current Floor :",k,"       Elevator's direction : down")
            print('Customers:')
            for i in range(self.customer_number):
                print(self.customer_name[i],'       current floor :',self.cur_list[i],'       destination floor :',self.dest_list[i])
                print('                 in Elevator :',self.in_elevator[i],'       is Finished :',self.get_out[i])
            print('\n------>------->-------->------->')
        print('\nAll Customers Arrived at Destination')
t=Elevator()                
```
