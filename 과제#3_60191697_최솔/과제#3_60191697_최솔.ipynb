```python
def my_minus(x,y) : #x-y를 구하는 함수를 프로그래밍하였다.
    return x-y
def my_multiply(x,y) : #x*y를 구하는 함수를 프로그래밍하였다.
    return x*y
def my_plus(x,y) : #x+y를 구하는 함수를 프로그래밍하였다.
    return x+y
def my_division(x,y) : #x/y를 구하는 함수를 프로그래밍하였다.
    return x/y
def my_calculate(x,y,Op) :
    #Op값에 따라 Op가 - 일때는 my_minus, 
    #Op가 * 일때는 my_multiply,
    #Op가 / 일때는 my_division,
    #Op가 + 일때는 my_plus 값을 출력하는 함수를 프로그래밍하였다. 
    if Op=="-" :           
        return my_minus(x,y)
    elif Op=="*" :
        return my_multiply(x,y)
    elif Op=="/" :
        return my_division(x,y)
    else :
        return my_plus(x,y)
def InputInt(s) :
    
    k=s 
    #k로 first를 설정한다. 
    
    s=input("Enter an integer:") 
    #넣고 싶은 값을 입력하고 이를 s로 둔다.
    
    def isInt(s): 
        #for 문을 이용해서 s 안의 문자가 "0123456789"에 없으면
        #isInt(s)값은 False를 return, 그렇지 않으면 True를 return 하는
        #함수 isInt(s)를 프로그래밍 한다.
        
        for i in range(0,len(s)) :            
            if s[i] not in"0123456789":
                return False
        return True
    
    while (isInt(s) is False) or len(s)==0  :
        print("Wrong integer! Input the",k,"integer again!" )
        s=input("Enter an integer:")
    return int(s)
#isInt(s)값이 False이거나 len(s)가 0일 경우, 
#while 문을 사용하여 그렇지 않을 때 까지 새로운 s값을 출력받는다.

def InputOp() :
    l=input("Enter an operation:")
    #l을 입력받은 값으로 저장한다.
    
    def IsOp(l):
        #조건문을 이용해 l이 "+-*/"에 해당하고, len(1)이 1이라면
        #True를 return하고, 그렇지 않으면 False를 return 하는
        #함수 IsOp(l)을 프로그래밍 한다.
        
        if l in "+-*/" and len(l)==1:
            return True
        else : return False
        
    while IsOp(l)==False :
        print("Wrong Operation! Input the Operation again!" )
        l=input("Enter an Operation:")
    return l
#while문을 사용하여 IsOp(l)이 False가 아닐 때까지
#새로운 l값을 출력받는다.
    
x=InputInt("first")
y=InputInt("second")
Op=InputOp()
print(str(x)+Op+str(y),"is",my_calculate(x,y,Op))
```
